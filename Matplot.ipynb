#### 1.Generar las siguiente figuras


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = np.arange(0,360)
radianes = np.radians(datos)
datos = np.sin(radianes)
plt.title("Funcin seno")
plt.xlabel("ngulo(rad)")
plt.ylabel("Sin(x)")
plt.plot(radianes,datos,"m--")
plt.show()
```

```python
datos = np.arange(0,360)
radianes = np.radians(datos)
datosSin = np.sin(radianes)
datosCos = np.cos(radianes)
plt.plot(radianes,datosSin,"m--",label="Funcin seno")
plt.plot(radianes,datosCos,"-g",label="Funcin coseno")
plt.xlabel("ngulo(rad)")
plt.legend(loc='upper right')
plt.show()
```

#### 2. Generar un histograma, como el que se muestra a continuacin, nusando los datos aleatorios con el metodo np.random.randn()


```python
data = np.random.randn(31)
x = np.linspace(0,30, endpoint=True)
plt.hist(data, bins=30, density=True, alpha=0.2, histtype='stepfilled', color="purple", edgecolor='none')
plt.plot(x,data)
```
