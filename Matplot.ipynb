#### 1.Generar las siguiente figuras


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = np.arange(0,360)
radianes = np.radians(datos)
datos = np.sin(radianes)
plt.title("Función seno")
plt.xlabel("Ángulo(rad)")
plt.ylabel("Sin(x)")
plt.plot(radianes,datos,"m--")
plt.show()
```

```python
datos = np.arange(0,360)
radianes = np.radians(datos)
datosSin = np.sin(radianes)
datosCos = np.cos(radianes)
plt.plot(radianes,datosSin,"m--",label="Función seno")
plt.plot(radianes,datosCos,"-g",label="Función coseno")
plt.xlabel("Ángulo(rad)")
plt.legend(loc='upper right')
plt.show()
```

#### 2. Generar un histograma, como el que se muestra a continuación, nusando los datos aleatorios con el metodo np.random.randn()


```python
mu, sigma = 20, 2
x = mu + sigma * np.random.randn(10000)
xr = np.arange(12,30,0.1)
plt.hist(x, bins=200, density=True, alpha=0.2, color="purple")
plt.plot(xr,(1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((xr - mu)/sigma)**2),"m--")
plt.xlim(12, 30)       
plt.ylim(0, 0.25)
plt.xlabel('Valor x')
plt.ylabel('Probabilidad')
plt.title('Histograma')
```

#### 3. Grafique las siguientes funciones en un mismo plano:
##### a) 2x+1
##### b) -x^2

```python
figura = plt.figure()
ax = plt.axes()
x = np.arange(-20,20,dtype=int)
ax.plot(x, (2*x+1), label='2x + 1')
ax.plot(x, (-x**2),label='-x^2')
ax.set_title("Funciones")
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.legend()
ax.grid()
ax.axhline(y=0, xmin=0, xmax=1,color="black")
ax.axvline(x=0, ymin=0, ymax=1,color="black")
```

#### 4. Genere un subplot con 6 paneles, usando la versión “a mano”, y con proporciones diferentes

```python
#Intento 1
x = np.arange(11)
y = 2*x
plt.figure() 
plt.subplot(2, 3, 1)
plt.plot(x,y)
plt.subplot(2, 3, 2)
plt.plot(x,y**2)
plt.subplot(2, 3, 3)
plt.plot(x,-y)
plt.subplot(2, 3, 4)
plt.plot(x,-y**2)
plt.subplot(2, 3, 5)
plt.plot(x,np.sin(y))
plt.subplot(2, 3, 6)
plt.plot(x,np.cos(y))
plt.subplots_adjust(hspace=0.4, wspace=0.4)
```
